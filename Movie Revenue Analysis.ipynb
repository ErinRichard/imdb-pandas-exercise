```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb-movies.csv')
```

### Drop columns without neccesary information and remove all records with no financial information -- Pay close attention to things that don't tell you anything regarding financial data

```python
# Columns to drop:  cast, homepage, tagline, keywords, overview, runtime, release_date, vote_count, vote_average
```

```python
# Make copy of dataframe (to maintain original)
imdb = df.copy(deep = True)
imdb
```

```python
# Drop Columns
# Need axis = 1 to specify columns, not rows
# Need inplace = True
imdb.drop(['cast', 'homepage', 'tagline', 'keywords', 'overview', 'runtime', 'release_date', 'vote_count', 'vote_average'], axis=1, inplace=True)
imdb.head()
```

```python
# drop popularity column
imdb.drop('popularity', axis=1, inplace=True)

# checking to see that it worked
imdb.head(1)
```

### Data Cleaning

```python
# Delete all records with null, or empty values
# To avoid skewing data
imdb_new = imdb.dropna()

# Confirm: Old number of rows 10866. New number of rows 9805
imdb_new
```

#### Here's a helpful hint from my own analysis when I ran this the first time. This may help shed light on what your data set should look like.

#### If I created one record for each the `production_companies` a movie was release under and one record each for `genres`<br>and tried to run calculations, it wouldn't work because for many records, the amount of `production_companies`<br>and `genres` aren't the same, so I'll create 2 dataframes; one w/o a `production_companies` column and one w/o a `genres` columns

```python
# Make copy of dataframe & delete production_companies column
no_prod = imdb_new.copy(deep = True)

no_prod.drop('production_companies', axis=1, inplace=True)

no_prod
```

```python
# Make copy of dataframe & delete genres column
no_genre = imdb_new.copy(deep = True)

no_genre.drop('genres', axis=1, inplace=True)

no_genre
```

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
# Determine which movies have been released within the last 10 years (True or False)
no_genre['movie_count'] = 2021 - no_genre['release_year'] <=10
no_genre.sort_values('movie_count', ascending = False).head()


# Create a count column for the count
no_genre['count'] = 1


# Filter on the movies released within the last 10 years ('movie_count' == True)
# Group by Production Company and sum the count of listings for each Production Company
ten_yrs = pd.DataFrame(no_genre[(no_genre['movie_count'] == True)].groupby('production_companies').sum()['count'])


print(f" The Asylum production company released the most movies (18) in the last 10 years:")

# Sort from highest to lowest count
# Dispay top five with .head()
ten_yrs.sort_values('count', ascending=False).head()


```
